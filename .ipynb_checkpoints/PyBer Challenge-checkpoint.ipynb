```python
# Loading and Reading CSV files
```

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
# Merge the DataFrames
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
## Deliverable 1.- Get a Summary DataFrame
```

```python
#1. Get the total rides for each city type
total_rides_per_city = pyber_data_df.groupby(["type"]).count()["ride_id"] 
total_rides_per_city
```

```python
# 2. Get the total drivers for each city type
drivers_per_city = city_data_df.groupby(["type"]).sum()["driver_count"] 
drivers_per_city
```

```python
# 3. Get the total amount of fares for each city type 
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_city = sum_fares_by_type / total_rides_per_city
avg_fare_per_city
```

```python
# 5. Get the average fare per driver for each city type.
avg_per_driver_by_city = sum_fares_by_type / drivers_per_city
avg_per_driver_by_city
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
        "Total Rides": total_rides_per_city,
        "Total Drivers":drivers_per_city,
        "Total Fares": sum_fares_by_type,
        "Average Fare per Ride": avg_fare_per_city, 
        "Average Fare per Driver": avg_per_driver_by_city})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = ""
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)
# Display the data frame
pyber_summary_df
```

```python
## Deliverable 2.- Create a multiple line plot that shows the total weekly of the fares for each type of city.
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_by_date_df = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
sum_fares_by_date_df.head()
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_fares_by_date_df = sum_fares_by_date_df.reset_index()
sum_fares_by_date_df.head()
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fares_by_date_df = sum_fares_by_date_df.pivot(index="date", columns="type", values="fare")
sum_fares_by_date_df.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_january_april_df = sum_fares_by_date_df.loc["2019-01-01":"2019-04-28"]
fares_january_april_df.head(10)
```

```python
fares_january_april_df.info()
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_january_april_df.index = pd.to_datetime(fares_january_april_df.index)
fares_january_april_df.head(10)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fares_january_april_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_per_week_df = fares_january_april_df.resample("W").sum()
fares_per_week_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
plt.style.use('fivethirtyeight')

# Graph the resampled DataFrame using OOI method
ax = fares_per_week_df.plot(figsize=(18, 6))
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare($)")
ax.set_xlabel("")
# Save the figure
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
